```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Sales dataset
data = {
"Month": ["Jan", "Feb", "Mar", "Jan", "Feb", "Mar", "Jan", "Feb", "Mar"],
"Product": ["Mobile", "Mobile", "Mobile", "Laptop", "Laptop", "Laptop", "TV",
"TV", "TV"],
"Units_Sold": [200, 220, 210, 120, 135, 150, 80, 95, 100],
"Revenue": [300000, 330000, 315000, 600000, 675000, 750000, 240000, 285000,
300000]
}
df = pd.DataFrame(data)
df
```

# Total revenue per product.

```python
Revresult = df.groupby("Product")["Revenue"].sum()
Revresult
```

# Average units sold per product

```python
Avgresult = df.groupby("Product")["Units_Sold"].mean()
Avgresult
```

# Monthly Revenue Trend for each product (Janâ€“Mar).

```python
import matplotlib.pyplot as plt
```

```python
monthlytrend = df.pivot(index='Month',columns='Product',values='Revenue')
monthlytrend.plot(marker='o')
plt.title("Monthly Revenue Trend")
plt.ylabel("Revenue")
plt.xlabel("Month")
plt.show()
```

# Bar chart of total revenue per product

```python
Revresult.plot(kind="bar")
plt.title("Total revenue per product")
plt.ylabel("Revenue")
plt.xlabel("Product")
plt.show()
```

# pie chart showing the contribution of each product to total revenue

```python
Revresult.plot(kind='pie')
plt.title("Contribution of each product to total revenue")
plt.show()
```
